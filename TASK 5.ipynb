```python
import pandas as pd
import numpy as np

df = pd.read_csv("/content/tested.csv")

df.head()
```

# Step 1: Dataset Overview
We loaded the Titanic dataset using pandas and checked the first few rows with `.head()`.  
This helped us understand the structure of the data and identify missing values.

```python
df.info()
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)

df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

df.drop(columns=['Cabin'], inplace=True)
```

# Step 2: Handling Missing Age
The `Age` column had missing values.  
We filled them with the **median age** to avoid skewing the data with extreme values.

# Step 3: Handling Missing Embarked
The `Embarked` column had missing values.  
We filled them with the **mode (most frequent value)** since itâ€™s categorical data.

# Step 4: Dropping Cabin
The `Cabin` column had too many missing values.  
We dropped it because it would not add reliable information to the analysis.



```python
print("Before:", df.shape)
df.drop_duplicates(inplace=True)
print("After:", df.shape)
```

# Step 5: Removing Duplicates
We used `.drop_duplicates()` to remove duplicate rows.  
This ensures the dataset is clean and avoids bias in analysis.

```python
df['PassengerId'] = df['PassengerId'].astype(int)
df['Age'] = df['Age'].astype(float)
df['Fare'] = df['Fare'].astype(float)
```

# Step 6: Converting Data Types
We converted `PassengerId` to integer and `Age` to float.  
Correct data types are important for accurate calculations and analysis.



```python

df['AgeCategory'] = pd.cut(df['Age'], bins=[0,12,18,60,100], labels=['Child','Teen','Adult','Senior'])


df['FareBand'] = pd.qcut(df['Fare'], 4, labels=['Low','Medium','High','Very High'])
```

```python
df.to_csv("cleaned_data.csv", index=False)
```

# Step 7: Creating Age Categories
We created a new column `AgeCategory` to group passengers into Child, Teen, Adult, and Senior.  
This helps in analyzing survival rates by age groups.

# Step 8: Creating Fare Bands
We created a new column `FareBand` using quartiles.  
This allows us to compare survival rates across different ticket price ranges.
